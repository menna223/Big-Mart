```python

t=data['Item_Fat_Content'].value_counts()
t.plot(kind='bar',color='skyblue')
plt.xlabel('Item_Fat_Content')
plt.ylabel('Count')
plt.title('Count of Item_Fat_Content')
plt.grid(False)
plt.show()
```

```python
g=data['Item_Type'].value_counts()
plt.figure(figsize=(25, 10))
g.plot(kind='bar',color='gray')
#train['Item_Type'].hist(color='purple', edgecolor='white')
plt.xlabel('Item_Type')
plt.ylabel('Count')
plt.title('Count number of Item_Type')
plt.grid(False)
plt.show()
```

```python
colors=['pink','white']
g6=data['Outlet_Type'].value_counts()
g6.plot(kind='pie',autopct='%1.1f%%',startangle=190,colors=colors)
plt.title('Count of outlet types')
plt.show()
```

```python
group3=data.groupby('Outlet_Size')['Item_Outlet_Sales'].sum()
print(group3)
group3.plot(kind='bar',color='gray')
plt.xlabel('outlet_size')
plt.ylabel('total sales')
plt.title('total sales by each product size' )
plt.show()
```

```python
group2 = data.groupby('Item_Fat_Content')['Item_Outlet_Sales'].sum()
plt.figure(figsize=(8, 5))
group2.plot(kind='bar', color='black')
plt.xlabel('Item Fat Content')
plt.ylabel('Item Outlet Sales')
plt.title('Total Item Outlet Sales by Fat Content')
plt.xticks(rotation=45)
plt.show()
```

```python
colors=['pink','gray','white']
g8=data.groupby('Outlet_Location_Type')['Item_Outlet_Sales'].sum()
g8.plot(kind='pie',autopct='%1.1f%%',startangle=140,colors=colors)

plt.title('Total Sales By outlet types ')
```

```python
g10=data.groupby('Item_Type')['Item_Visibility'].sum()
g10.plot(kind='bar',color='skyblue')
plt.xlabel('Item_Type')
plt.ylabel('Visibility of each product')
plt.title('Visibility by product type')
plt.show()
```

```python
group1 = data.groupby('Item_Type')['Item_Outlet_Sales'].sum()

plt.figure(figsize=(10, 6))
group1.plot(kind='bar',color='pink')
plt.xlabel('Item_Type')
plt.ylabel('total sales of each product')
plt.title('Total sales by product type')
plt.show()
```

```python
group5=data.groupby('Outlet_Establishment_Year')['Item_Outlet_Sales'].sum()
print(group5)
group5.plot(kind='line',color='black')
plt.xlabel('Year')
plt.ylabel('total sales of product')

plt.title('Total sales by year')
plt.show()
```

```python
colors=['pink','gray','white']
g7=data.groupby('Outlet_Type')['Item_Outlet_Sales'].sum()
print(g7)
g7.plot(kind='pie',autopct='%1.1f%%',startangle=140,colors=colors)
plt.title('Total Sales by outlet type ')
```

```python
sns.pairplot(data,x_vars=['Item_Fat_Content','Item_Visibility','Item_Type','Item_MRP'],y_vars=['Item_Outlet_Sales'],hue='Item_Fat_Content')
```
